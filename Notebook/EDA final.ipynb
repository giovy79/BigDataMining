```
%sql
--- UNIFICANDO DATASETS
CREATE TABLE MovieLens AS
  SELECT r.userID, r.rating, m.*  FROM default.ratings r 
  JOIN default.movies m  
  ON r.movieId = m.movieId
```

```
%sql
--- NUMERO DE CALIFICACIONES POR USUARIO
SELECT num_calificaciones, COUNT(*) AS num_usuarios
FROM (
    SELECT userId, COUNT(*) AS num_calificaciones
    FROM default.MovieLens
    GROUP BY userId
) user_evaluaciones
GROUP BY num_calificaciones
ORDER BY num_calificaciones;
```

```
%sql
--- PROMEDIO DE CALIFACIONES POR USUARIO
SELECT AVG(cnt_ratings) AS promedio_calificaciones_por_usuario
FROM (
    SELECT userId, COUNT(*) AS cnt_ratings
    FROM default.MovieLens
    GROUP BY userId
) user_ratings;
```

```
%sql
--- NUMERO DE CALIFICACIONES POR PELICULA
SELECT num_calificaciones, COUNT(*) AS num_peliculas
FROM (
    SELECT movieId, COUNT(*) AS num_calificaciones
    FROM default.MovieLens
    GROUP BY movieId
) movie_evaluaciones
GROUP BY num_calificaciones
ORDER BY num_calificaciones;
```

```
%sql
--- PROMEDIO DE CALIFACIONES POR PELICULA
SELECT AVG(cnt_ratings) AS promedio_calificaciones_por_pelicula
FROM (
    SELECT movieId, COUNT(*) AS cnt_ratings
    FROM  default.MovieLens
    GROUP BY movieId
) movie_ratings;
```

```
%sql
--- CANTIDAD DE CALIFICACIONES SEGUN LOS PUNTAJES
SELECT
    puntaje,
    COUNT(*) AS cantidad_calificaciones
FROM (
    SELECT 0 AS puntaje UNION ALL
    SELECT 0.5 UNION ALL
    SELECT 1 UNION ALL
    SELECT 1.5 UNION ALL
    SELECT 2 UNION ALL
    SELECT 2.5 UNION ALL
    SELECT 3 UNION ALL
    SELECT 3.5 UNION ALL
    SELECT 4 UNION ALL
    SELECT 4.5 UNION ALL
    SELECT 5
) AS puntajes
LEFT JOIN default.MovieLens ON puntajes.puntaje = default.MovieLens.rating
GROUP BY
    puntaje
ORDER BY
    puntaje
```

```
%sql
--- CANTIDAD DE PELICULAS SEGUN EL PROMEDIO DE LAS CALIFACIONES
SELECT
    promedio_calificacion,
    COUNT(*) AS cantidad_peliculas
FROM (
    SELECT
        movieId,
        ROUND(AVG(rating), 0) AS promedio_calificacion
    FROM
        default.MovieLens
    GROUP BY
        movieId
) subquery
GROUP BY
    promedio_calificacion
ORDER BY
    promedio_calificacion;
```

```
%sql
-- Crear una tabla temporal con los valores de calificación deseados
CREATE OR REPLACE TEMPORARY VIEW valores_calificacion AS
SELECT 0 AS puntaje UNION ALL
SELECT 0.5 UNION ALL
SELECT 1 UNION ALL
SELECT 1.5 UNION ALL
SELECT 2 UNION ALL
SELECT 2.5 UNION ALL
SELECT 3 UNION ALL
SELECT 3.5 UNION ALL
SELECT 4 UNION ALL
SELECT 4.5 UNION ALL
SELECT 5;

-- Consulta para contar la cantidad de películas según los valores de calificación
SELECT
    v.puntaje AS promedio_calificacion,
    COUNT(s.movieId) AS cantidad_peliculas
FROM
    valores_calificacion v
LEFT JOIN (
    SELECT
        movieId,
        ROUND(AVG(rating), 1) AS promedio_calificacion
    FROM
        default.MovieLens
    GROUP BY
        movieId
) s
ON
    v.puntaje = s.promedio_calificacion
GROUP BY
    v.puntaje
ORDER BY
    v.puntaje;
```

```
%sql
SELECT genre, COUNT(*) AS cantidad_peliculas
FROM (
    SELECT DISTINCT movieId, 
           TRIM(SUBSTRING_INDEX(SUBSTRING_INDEX(genres, '|', n), '|', -1)) AS genre
    FROM default.MovieLens
    JOIN (
        SELECT 1 AS n UNION ALL
        SELECT 2 UNION ALL
        SELECT 3 UNION ALL
        SELECT 4 UNION ALL
        SELECT 5 UNION ALL
        SELECT 6 UNION ALL
        SELECT 7 UNION ALL
        SELECT 8 UNION ALL
        SELECT 9 UNION ALL
        SELECT 10
    ) numbers
    ON CHAR_LENGTH(genres) - CHAR_LENGTH(REPLACE(genres, '|', '')) >= n - 1
) movie_genres
GROUP BY genre
ORDER BY cantidad_peliculas DESC;
```
